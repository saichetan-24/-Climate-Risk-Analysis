```python
num_cols = ['Temp', 'D.O. (mg/l)', 'PH',
            'CONDUCTIVITY (¬µmhos/cm)', 'B.O.D. (mg/l)',
            'NITRATENAN N+ NITRITENANN (mg/l)',
            'FECAL COLIFORM (MPN/100ml)',
            'TOTAL COLIFORM (MPN/100ml)Mean']

for col in num_cols:
    sns.histplot(df_water[col], kde=True)
    plt.title(f"{col} Distribution")
    plt.show()
```

```python
# force conversion of numeric columns, turn errors into NaN
for col in num_cols:
    df_water[col] = pd.to_numeric(df_water[col], errors='coerce')
```

```python
df_water[num_cols] = df_water[num_cols].fillna(df_water[num_cols].mean())
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_water[num_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap - Water Quality Parameters")
plt.show()
```

```python
df_merged = pd.merge(df_rain, df_water, left_on=['STATE_UT_NAME', 'ANNUAL'],
                     right_on=['STATE', 'year'], how='inner')

print("‚úÖ Merged Dataset Shape:", df_merged.shape)
df_merged.head()
```

```python
# Fill missing values with column mean
df_merged.fillna(df_merged.mean(), inplace=True)
```

```python
df_merged = pd.get_dummies(df_merged, drop_first=True)
```

```python
print("Rainfall dataset shape:", df_rain.shape)
print("Water dataset shape:", df_water.shape)
```

```python
df_merged = pd.merge(
    df_rain, df_water,
    left_on="STATE_UT_NAME",
    right_on="STATE",
    how="inner"
)

print("‚úÖ Merged Dataset Shape:", df_merged.shape)
df_merged.head()
```

```python
# Replace blanks with NaN and convert to float
for col in df_water.columns:
    df_water[col] = pd.to_numeric(df_water[col], errors="ignore")

# Re-check merged numeric columns
num_df = df_merged.select_dtypes(include="number")
print("Numeric columns:", num_df.columns.tolist())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(num_df)

df_scaled = pd.DataFrame(scaled_data, columns=num_df.columns)
df_scaled.head()
```

```python
# festure selection
X = df_merged.select_dtypes(include='number').drop("ANNUAL", axis=1)  # features
y = df_merged["ANNUAL"]  # target (rainfall)
```

```python
from sklearn.feature_selection import SelectKBest, f_regression
import pandas as pd

# Select top 10 features
selector = SelectKBest(score_func=f_regression, k=10)
fit = selector.fit(X, y)

# Store results in dataframe
feature_scores = pd.DataFrame({
    "Feature": X.columns,
    "Score": fit.scores_
}).sort_values(by="Score", ascending=False)

print("üîù Top 10 Features influencing Rainfall:")
display(feature_scores.head(10))
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    data=feature_scores.head(10),
    x="Score",
    y="Feature",
    hue="Feature",
    palette="viridis",
    dodge=False,
    legend=False
)
plt.title("Top 10 Features Influencing Rainfall")
plt.show()
```
